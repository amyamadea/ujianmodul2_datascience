### 5. Make Table like this to describe the dataframe

#### Columns Explanations :
 - __dataFeatures__ = column name
 - __dataType__ = data type of that column
 - __null__ = how many null data in that column/series
 - __nullPct__ = null percentage of total data in that column
 - __unique__ = how many unique data
 - __uniqueSample__ = get some unique sample from each column

#### hint : search .sample(), .drop_duplicates(), .values, .isna(), dtype

```python
data = []
countData = len(titanicDF)

for i in range(len(titanicDF.columns)):
    countNull = len(titanicDF[titanicDF[titanicDF.columns[i]].isnull()])
    nullPct = countNull/countData * 100
    
    sample = []
    for item, x in zip(titanicDF[titanicDF.columns[i]].drop_duplicates().sample(n=2), range(2)):
        sample.append(item)
    
    data.append([
        titanicDF.columns[i],
        titanicDF[titanicDF.columns[i]].dtypes, 
        countNull, 
        nullPct, 
        titanicDF[titanicDF.columns[i]].nunique(), 
        sample
    ])
    
describeDF = pd.DataFrame(columns=['dataFeatures', 'dataType', 'null', 'nullPct', 'unique', 'uniqueSample'], data=data)
describeDF
```

### 8. Missing Values
#### Berapa banyak NaN age data dan mean dari setiap who?
#### Isi data NaN dari kolom age dengan mean dari umur tergantung whonya dia apa (child/woman/man)
#### Kemudian print banyak NaN age data dan mean umur setiap who lagi.

```python
def printNanMean():
    for i in range(titanicDF['who'].nunique()):
        isNull = titanicDF[(titanicDF['who']==titanicDF['who'].unique()[i]) & (titanicDF['age'].isnull())]
        countNull = len(isNull)
        mean = titanicDF[(titanicDF['who']==titanicDF['who'].unique()[i])]['age'].mean()

        print ('Banyak NaN ' + titanicDF['who'].unique()[i] + ' & mean: ' + str(countNull) + ' & ' + str(mean))

printNanMean()
```

```python
# get mean age per sex
ageMeanFemale = titanicDF.groupby(['sex']).mean()['age']['female']
ageMeanMale = titanicDF.groupby(['sex']).mean()['age']['male']

for i in range(len(titanicDF['age'])):
    if (titanicDF['age'].isnull()[i]):
        if (titanicDF['sex'][i]=='male'):
            titanicDF['age'][i] = ageMeanMale
        elif (titanicDF['sex'][i]=='female'):
            titanicDF['age'][i] = ageMeanFemale

# def setAge(age):
#     if (math.isnan(age)==True):
#         return 99
#     else:
#         return age

# titanicDF['age'] = titanicDF['age'].apply(lambda age: setAge(age))
# titanicDF.head(10)
```

#### Isi NaN embarked dan embark_town dan deck dengan modus masing2 (print modus masing2 juga)

```python
def getModulus(column):
    modDF = titanicDF.head()
    value_counts = titanicDF[column].value_counts(dropna=True, sort=True)
    modDF = value_counts.rename_axis('unique_values').reset_index(name='counts')
    
    return modDF[modDF['counts'] == modDF['counts'].max()]['unique_values'][0]
```

```python
titanicDF['embarked'].fillna(getModulus('embarked'), inplace=True)
print('Modus Embarked = ' + getModulus('embarked'))
```

```python
titanicDF['embark_town'].fillna(getModulus('embark_town'), inplace=True)
print('Modus Embark Town  = ' + getModulus('embark_town'))
```

```python
titanicDF['deck'].fillna(getModulus('deck'), inplace=True)
print('Modus Deck  = ' + getModulus('deck'))
```

### 9. Adding New Features/Columns/Variables

#### Buat DataFrame baru dengan column2 sebagai berikut
 - __id__ = id table ini (PK)
 - __mean_fare__ = Mean Fare
 - __std_fare__ = Std Fare
 - __batas_min_fare__ = Mean Fare - (3 x Std Fare)
 - __batas_max_fare__ = Mean Fare + (3 x Std Fare)
 
#### Dan data2nya ada 3 (fare setiap class yaitu first, second, and third class)

```python
mean_fare = titanicDF.groupby(['class']).mean()['fare']
std_fare = titanicDF.groupby(['class']).std()['fare']

data = []

for i in range(len(titanicDF['class'].unique())):
    data.append([
        i+1,
        mean_fare[titanicDF['class'].unique()[i]],
        std_fare[titanicDF['class'].unique()[i]],
        mean_fare[titanicDF['class'].unique()[i]] - (3 * std_fare[titanicDF['class'].unique()[i]]),
        mean_fare[titanicDF['class'].unique()[i]] + (3 * std_fare[titanicDF['class'].unique()[i]])
    ])

    
titanicCalcDF = pd.DataFrame(columns=['id', 'mean_fare', 'std_fare', 'batas_min_fare', 'batas_max_fare'], data=data)
print (titanicDF['class'].unique())
titanicCalcDF
```

#### Print outlier % of each class

```python
# titanicDF['class'].unique()

classDict = {'First': 0, 'Second': 0, 'Third': 0}

for i in range(len(titanicDF['class'].unique())):
    batas_min_fare = titanicCalcDF.iloc[i]['batas_min_fare']
    batas_max_fare = titanicCalcDF.iloc[i]['batas_max_fare']

    # cari jumlah outlier masing-masing class
    countOutlier = len(titanicDF[(titanicDF['class']==titanicDF['class'].unique()[i]) & ((titanicDF['fare']<batas_min_fare) | (titanicDF['fare']>batas_max_fare))])
    countClass = len(titanicDF[titanicDF['class']==titanicDF['class'].unique()[i]])

    # get presentase
    classDict[titanicDF['class'].unique()[i]] = countOutlier/countClass * 100
    

for key, value in classDict.items():
    print('Outlier % ' + key + ' : ' + str(value))
```

#### Tambahkan column outliercalcId (FK) di dfTitanic

```python
def setOutlierId(c):
    for i in range(len(titanicDF['class'].unique())):
        if (c==titanicDF['class'].unique()[i]):
            return i+1

titanicDF['outliercalcId'] = titanicDF['class'].apply(lambda c: setOutlierId(c))
titanicDF.head()
```

#### Tambahkan column outlier (1 for yes and 0 for no)

```python
for i in range(len(titanicDF['outliercalcId'])):
    batas_min_fare = titanicCalcDF[titanicCalcDF['id']==titanicDF['outliercalcId'][i]]['batas_min_fare'].iloc[0]
    batas_max_fare = titanicCalcDF[titanicCalcDF['id']==titanicDF['outliercalcId'][i]]['batas_max_fare'].iloc[0]
    
    if ((titanicDF['fare'][i]<batas_min_fare) | (titanicDF['fare'][i]>batas_min_fare)):
        titanicDF['outlier'] = 1
    else:
        titanicDF['outlier'] = 0

titanicDF.head()
```
